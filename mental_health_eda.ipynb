```python
# Correlation heatmap for numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

if len(numeric_cols) > 0:
    corr = df[numeric_cols].corr()
    plt.figure(figsize=(8, 6))
    im = plt.imshow(corr, interpolation="nearest")
    plt.title("Correlation matrix of numeric features")
    plt.colorbar(im, fraction=0.046, pad=0.04)
    plt.xticks(range(len(numeric_cols)), numeric_cols, rotation=90)
    plt.yticks(range(len(numeric_cols)), numeric_cols)
    plt.tight_layout()
    plt.show()
else:
    print("No numeric columns found for correlation matrix.")
```

```python
# Time-based line plot of mental health scores (if date exists)
if "date" in df.columns:
    df["date"] = pd.to_datetime(df["date"], errors="coerce")
    df_sorted = df.sort_values("date")
    time_cols = [c for c in ["stress_level", "anxiety_level", "mood_level"] if c in df.columns]
    if len(time_cols) > 0:
        plt.figure(figsize=(10, 5))
        for c in time_cols:
            plt.plot(df_sorted["date"], df_sorted[c], label=c)
        plt.title("Mental health scores over time")
        plt.xlabel("Date")
        plt.ylabel("Score")
        plt.legend()
        plt.tight_layout()
        plt.show()
    else:
        print("No mental health score columns found for time-based plot.")
else:
    print("Column 'date' not found.")
```
